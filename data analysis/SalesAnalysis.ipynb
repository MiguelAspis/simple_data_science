```python
df = pd.read_csv("../datasets/SalesAnalysis/Sales_Data/all_data.csv")
df['Months'] = df['Order Date'].str[0:2]
```

```python
df.head()
```

```python
df['Months'] = df['Months'].astype('int')
```

```python
df = df.dropna(how='all')
```

```python
df.head()
```

```python
df['Months'] = df['Months'].astype('int32')
```

```python
df = df[df['Order Date'].str[0:2] != 'Or']
```

```python
df['Months'] = df['Months'].astype('int32')
```

```python
df.head()
```

```python
df['Sales'] = df['Quantity Ordered']*df['Price Each']
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])

df['Sales'] = df['Quantity Ordered']*df['Price Each']
```

```python
df.head()
```

```python
sales_total = df.groupby('Months').sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months,sales_total['Sales'])
plt.xticks(months)
plt.ylabel("Sales in USD ($)")
plt.xlabel("Months")
```

```python
df['City'] = df['Purchase Address'].apply(lambda x: x.split(',')[1] + ' ' + x.split(',')[2].split(' ')[1])
```

```python
df.head()
```

```python
city_sales_total = df.groupby('City').sum()
```

```python
city_sales_total
```

```python
import matplotlib.pyplot as plt

cities = [city for city, df in df.groupby('City')]

plt.bar(cities,city_sales_total['Sales'])
plt.xticks(cities, rotation='vertical',size=8)
plt.ylabel("Sales in USD ($)")
plt.xlabel("City Name")
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.head()
```

```python
df['Hour'] = df['Order Date'].dt.hour
```

```python
df.head()
```

```python
df['Minute'] = df['Order Date'].dt.minute
```

```python
df.head()
```

```python
hours = [hour for hour, df in df.groupby('Hour')]

plt.plot(hours, df.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hours')
plt.grid()
plt.ylabel('Number of Orders')
```

```python
new_df = df[df['Order ID'].duplicated(keep=False)]

new_df['Grouped'] = new_df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

new_df = new_df[['Order ID', 'Grouped']].drop_duplicates()
```

```python
new_df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in new_df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
count.most_common(10)
```

```python
product_group = df.groupby('Product')

most_sold = product_group.sum()['Quantity Ordered']

products = [product for product, df in product_group]

plt.bar(products, most_sold)
plt.xticks(products, rotation='vertical',size=8)
plt.show()
```

```python
prices = df.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, most_sold, color='g')
ax2.plot(products,prices, 'b-')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Most Sold Products', color='g')
ax2.set_ylabel('Price', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)
plt.show()
```
